# 01 — Gradient Check

Verifica che il gradiente analitico di $f(w)$ sia corretto,
confrontandolo con le differenze finite centrali.

$$f(w) = \frac{1}{2}\|X^T w - y\|^2 + \frac{1}{2}\lambda^2 \|w\|^2$$

```python
import sys
sys.path.insert(0, '../..')

import numpy as np
from utils import compute_loss, check_gradient, load_ml_cup, solve_exact
```

## Test su un set di dati casuale

```python
np.random.seed(42)
m = 50  # righe tall
n= 10   # colonne (thin)
lam = 0.5

X = np.random.randn(m, n)
y = np.random.randn(n)
w = np.random.randn(m)

print(f"Test su dimensioni: X={X.shape}, w={w.shape}, y={y.shape}")

# Test Funzione Obiettivo
loss_val = compute_loss(w, X, y, lam)
print(f"Valore f(w): {loss_val:.4f}")

# Test gradiente
abs_err, rel_err = check_gradient(w, X, y, lam)
print(f"Errore assoluto: {abs_err:.2e}")
print(f"Errore relativo: {rel_err:.2e}")

assert rel_err < 1e-5, "Gradiente non corretto!"
print("✅ Gradiente corretto!")
```

### Test su diversi valori di $\lambda$

```python
for lam_test in [1e-4, 1e-2, 0.5, 1.0, 10.0, 1e4]:
    abs_e, rel_e = check_gradient(w, X, y, lam_test)
    status = '✅' if rel_e < 1e-5 else '❌'
    print(f"  λ={lam_test:<8}  rel_err={rel_e:.2e}  {status}")
```

### Test su dati reali (ML-CUP)

```python
file_path = r"Dataset\ML-CUP25-TR.csv" # Assicurati che il percorso sia giusto
np.random.seed(42)
lam = 0.5

# Carica i dati reali
X, y, m, n = load_ml_cup(file_path)
print(f"Dataset caricato: m={m}, n={n}")

# Inizializza w casualmente
w = np.random.randn(m)

# CALCOLO DELLA BASELINE (Soluzione Esatta)
print("\n--- CALCOLO BASELINE (Equazioni Normali) ---")
w_baseline, f_baseline, tempo_baseline = solve_exact(X, y, lam)

print(f"Tempo calcolo Baseline: {tempo_baseline:.5f} sec")
print(f"Peso w* (prima 5 componenti): {w_baseline[:5]}")
print(f"Valore minimo ESATTO f(w*): {f_baseline:.8f}")

# Test Funzione Obiettivo
loss_val = compute_loss(w, X, y, lam)
print(f"Valore f(w) iniziale: {loss_val:.4f}")

# Test Gradiente
abs_err, rel_err = check_gradient(w, X, y, lam)
print(f"Errore assoluto: {abs_err:.2e}")
print(f"Errore relativo: {rel_err:.2e}")

assert rel_err < 1e-5, "Gradiente non corretto!"
print("✅ Gradiente corretto!")
```

### Test su diversi valori di $\lambda$ (ML-CUP)

```python
for lam_test in [1e-4, 1e-2, 0.5, 1.0, 10.0, 1e4]:
    abs_e, rel_e = check_gradient(w, X, y, lam_test)
    status = '✅' if rel_e < 1e-5 else '❌'
    print(f"  λ={lam_test:<8}  rel_err={rel_e:.2e}  {status}")
```
