# 01 — Gradient Check

Verify that the analytical gradient of $f(w)$ is correct by comparing it with central finite differences.

$$f(w) = \frac{1}{2}\|X^T w - y\|^2 + \frac{1}{2}\lambda^2 \|w\|^2$$


```python
import sys
sys.path.insert(0, '../..')

import numpy as np
from utils import compute_loss, check_gradient, load_ml_cup, solve_exact
```

## Test on random dataset

```python
np.random.seed(42)
m = 50  # rows (tall)
n = 10  # columns (thin)
lam = 0.5

X = np.random.randn(m, n)
y = np.random.randn(n)
w = np.random.randn(m)

print(f"Testing dimensions: X={X.shape}, w={w.shape}, y={y.shape}")

# Test Objective Function
loss_val = compute_loss(w, X, y, lam)
print(f"Value of f(w): {loss_val:.4f}")

# Test gradient
abs_err, rel_err = check_gradient(w, X, y, lam)
print(f"Absolute error: {abs_err:.2e}")
print(f"Relative error: {rel_err:.2e}")

assert rel_err < 1e-5, "Incorrect gradient!"
print("✅ Gradient is correct!")
```

### Test on different values of $\lambda$

```python
for lam_test in [1e-4, 1e-2, 0.5, 1.0, 10.0, 1e4]:
    abs_e, rel_e = check_gradient(w, X, y, lam_test)
    status = '✅' if rel_e < 1e-5 else '❌'
    print(f"  λ={lam_test:<8}  rel_err={rel_e:.2e}  {status}")
```

### Test on real data (ML-CUP)

```python
file_path = r"Dataset\ML-CUP25-TR.csv"
np.random.seed(42)
lam = 0.5

# Load the real data
X, y, m, n = load_ml_cup(file_path)
print(f"Dataset loaded: m={m}, n={n}")

# Initialize w randomly
w = np.random.randn(m)

# BASELINE CALCULATION (Exact Solution)
print("\n--- BASELINE CALCULATION (Normal Equations) ---")
w_baseline, f_baseline, baseline_time = solve_exact(X, y, lam)

print(f"Baseline calculation time: {baseline_time:.5f} sec")
print(f"Weight w* (first 5 components): {w_baseline[:5]}")
print(f"EXACT minimum value f(w*): {f_baseline:.8f}")

# Test Objective Function
loss_val = compute_loss(w, X, y, lam)
print(f"Initial f(w) value: {loss_val:.4f}")

# Test Gradient
abs_err, rel_err = check_gradient(w, X, y, lam)
print(f"Absolute error: {abs_err:.2e}")
print(f"Relative error: {rel_err:.2e}")

assert rel_err < 1e-5, "Incorrect gradient!"
print("✅ Gradient is correct!")
```

### Test on different values of $\lambda$ (ML-CUP)

```python
for lam_test in [1e-4, 1e-2, 0.5, 1.0, 10.0, 1e4]:
    abs_e, rel_e = check_gradient(w, X, y, lam_test)
    status = '✅' if rel_e < 1e-5 else '❌'
    print(f"  λ={lam_test:<8}  rel_err={rel_e:.2e}  {status}")
```
